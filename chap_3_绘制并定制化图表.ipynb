# 直方图 的绘制
## 调用 matplotlib.pyplot.hist() 来创建 直方图 需要传入一些参数
1. bins: 可以是一个 bin 数量的整数值 亦可是表示 bin 的序列 默认值是 10
2. range: bin 的范围 当 bins 参数为序列时 此参数无效 范围外的值将被忽略掉 默认值为 None
3. normed: 若值为 True 直方图的值将 归一化 处理 形成概率密度 默认值为 False
4. histtype: 默认值为 bar 类型的直方图 还有其他类型
    - barstacked： 用于多种数据的堆叠直方图
    - step: 创建为填充的线形图
    - stepfilled: 创建默认填充的线形图
5. align: 用于 bin 边界之间矩形条的居中设置 默认值为 mid 其他值为 left AND right
6. color: 指定直方图颜色 可以是单一颜色 OR 颜色的序列 IF 指定了多个数据集合 颜色序列会设置为相同的顺序 IF 未指定 将会指定使用默认的颜色
7. orientation: 通过指定 orientation 为 horizontal 创建 水平 直方图 默认是 vertical

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 10000)
ax = plt.gca()

# the histogram of the data
ax.hist(x, bins=35,color='r')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title(r'$\mathrm{Histogram:}\ \mu=%d,\ \sigma=%d$' % (mu, sigma))
plt.show()
```

# 误差条形图的绘制
## 利用误差条来可视化数据集合中的测量不确定度 OR 指出错误 误差条可很容易表示出误差偏离数据集合的情况 他们可显示在一个标准差 一个标准误差 OR 95% 的置信区间
### 必选 argument left AND right 可选 argument
1. width: 给定误差条的宽度 默认值为 0.9
2. bottom: IF 指定 bottom 其值会加到高度中去 默认值为 None
3. edgecolr: 给定误差条边界颜色
4. linewidth：误差条边界宽度 可设置为 None AND 0(此时误差条边界将不显示)
5. orientation: 有 vertical AND horizontal
6. xerr AND yerr: 用于在柱状图上生成误差条
以上可选 arguments 可以是 单一值 亦可是 和误差条数目相同长度的序列

```python
import numpy as np
import matplotlib.pyplot as plt

# generate number of measurements
x = np.arange(0,10,1)

# valueds computed from 'measured'
y = np.log(x)

# add some error samples from standard normal distribution
xe = 0.1 * np.abs(np.random.randn(len(y)))

# draw and show errobar
plt.bar(x,y,yerr=xe,width=0.4,align='center',ecolor='r',color='cyan',label='experiment # 1')
# give some explaintions
plt.xlabel('# measurement')
plt.ylabel('Measured values')
plt.title('Measurements')
plt.legend(loc='upper left')
plt.show()
```

# 绘制饼图
## 显示数据集合加起来必须等于 100%
### 饼图描述数值 比例关系 其中每个扇区 弧长 大小为其所表示的数量的比例
#### 通过给定一个分裂序列 可获得分裂效果 其中每个元素表示圆弧之间偏移量为半径的百分比 用 autopct  参数 来格式化绘制在圆弧中的标签 标签可是一个格式化字符串 OR 可调用 的对象 （函数） AND 可使用 一个布尔值的阴影参数来给 饼图 添加 阴影效果<br>IF 未指定 startangle 扇区 将会从 x 轴 （angle = 0） 开始 逆时针排列  若指定则从 对应角度 开始排列

```python
from pylab import *

# make a square and axes
figure(1, figsize=(6, 6))
ax = axes([.1, .1, .8, .8])

# the slices will be ordered
# and plotted counter-clockwise
labels = 'Sqring', 'Summer', 'Autumn', 'Winter'

# fractions are either x/sum(x) or x if sum(x) <= 1
x = [155, 30, 45, 10]

# explode must be len(x) sequence or None
explode = (.1, .1, .1, .1)

pie(x, explode=explode, labels=labels, autopct='%1.1f%%', startangle=67)

title('Rainy days be season')
show()
```

# 绘制带填充区域的图表
## fill_between() method 使用 x 为定点选取 y 值 (y1,y2) 后将用预定义的颜色 绘制 多边形<br>用 where 参数 来指定一个条件来填充 曲线 where 参数 接受 布尔值 OR Expression 这样即可填充 where 条件区域
### fill_between 接受 许多参数 OR hatch (指定填充的样式替代颜色) AND 线条选项 (linewidth AND linestyle)<br>fill_betweenx() 仅仅针对水平曲线的填充特性<br>更加通用 fill() 提供对任意多边形填充颜色 OR 阴影线的功能

```python
'''
绘制填充区域的图表
'''

from matplotlib.pyplot import figure, show, gca
import numpy as np

x = np.arange(0.0,2,.01)
# two diff signals are measured
y1 = np.sin(2*np.pi*x)
y2 = 1.2*np.sin(4*np.pi*x)

fig = figure()
ax = gca()

# plot and fill between y1 and y2 where a logical condition is met
ax.plot(x,y1,x,y2,color='black')

ax.fill_between(x,y1,y2,where=y2>=y1,facecolor='darkblue',interpolate=True)
ax.fill_between(x,y1,y2,where=y2<=y1,facecolor='deeppink',interpolate=True)

ax.set_title('filled between')
show()
```

# 绘制带色彩标记的散点图
## 散点图 显示 两组数据 值
1. 设置图形颜色 color
2. 设置点状标记(默认 circle) marker alpha(透明度) edgecolors(标记边界颜色) label(图例框)

### 散点图 通常 应用于 拟合回归函数 用来识别两个 var 关联

```python
import matplotlib.pyplot as plt
import numpy as np

# generate x va
x = np.random.randn(10000)

# random measurements no correlation
y1 = np.random.randn(len(x))

# stong correlation
y2 = 1.2 + np.exp(x)

ax1 = plt.subplot(121)
plt.scatter(x, y1, color='indigo', alpha=0.3,
            edgecolors='white', label='no correl')
plt.xlabel('no correlation')
plt.grid(True)
plt.legend()

ax2 = plt.subplot(122, sharex=ax1, sharey=ax1)
plt.scatter(x, y2, color='green', alpha=0.3, edgecolors='grey', label='correl')
plt.xlabel('strong correlation')
plt.grid(True)
plt.legend()

plt.show()
```
